```python
import zipfile

with zipfile.ZipFile("student.zip", "r") as zip_ref:
    zip_ref.extractall("student_data")
import os
os.listdir("student_data")
import pandas as pd

df = pd.read_csv("student_data/student-mat.csv", sep=';')
df.head()
df.shape
df.info()
df.isnull().sum()
df = df.drop_duplicates()
df['G3'].mean()
df[df['G3'] > 15].shape[0]
df['studytime'].corr(df['G3'])
df.groupby('sex')['G3'].mean()
import matplotlib.pyplot as plt
plt.hist(df['G3'], bins=10)
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Grade by Gender")
plt.show()
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Grade by Gender")
plt.show()

```

## Conclusion

The analysis of the student performance dataset shows how different factors affect final grades.

The distribution of final grades (G3) shows that most students scored in the medium range, while a smaller number achieved very high or very low marks.

The scatter plot between study time and final grade indicates that students who spend more time studying generally tend to score better, showing a positive relationship between study time and performance.

The gender-wise bar chart shows a difference in the average performance between male and female students.
